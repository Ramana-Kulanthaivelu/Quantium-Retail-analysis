Checking for outliers using boxplot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'data'

# Select numerical columns
numerical_columns = data.select_dtypes(include='number').columns

# Create box plots for each numerical column to identify outliers
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=data[column])
    plt.title(f'Box plot of {column}')
    plt.show()
```

```python
import pandas as pd
from scipy.stats import zscore

# Calculate Z-score for each numerical column
z_scores = zscore(data.select_dtypes(include='number'))

# Define a threshold for identifying outliers 
threshold = 3

# Find outliers by checking if Z-score exceeds the threshold
outliers_mask = (z_scores > threshold) | (z_scores < -threshold)

# Remove outliers
cleaned_data = data[~outliers_mask.any(axis=1)]

#create a DataFrame with outliers only
outliers = data[outliers_mask.any(axis=1)]
```
